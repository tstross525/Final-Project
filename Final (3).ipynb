Introduction
------------
For our final project, our team has decided to measure the overall quality of life in Pittsburgh neighborhoods by combining several key factors:

1. Crime Rate: Lower crime is assumed to indicate a better living environment (Trent).

2. Housing: Factors such as housing affordability, quality, and availability (Richie).

3. Fire Incidents: Safety.

Crime
-----------
The dataset was used to determine the number of arrests from 1998-03-11 to 2023-11-14 in the city of Pittsburgh. Since there were several neighborhoods and tons of data, I filtered out only "Shadyside" in the code. My findings showed only 1.5% of arrests in the city of Pittsburgh occured in Shadyside.


Housing
-----------

The dataset was used to look at the housing statistics in the Pittsburgh area. It looked at how many available units were occupied and if people were renting vs buying. It also took into account living situations such as families vs other multiple-person living arrangements. Taking all of this data into account, a special statistic was created. Shadyside, the neighborhood we chose, ranked 14th with a .75, which is above the cutoff for a good neighborhood (25th or a .7).

Fire Incidents
--------------

The dataset was used to determine the percentage of fires from (08/04/2017-04/07/25) that occured in Shadyside compared to other neighborhoods in the city of Pittsburgh. Through my code the percentage of fires in Shadyside is 1.86% out of all the other neighborhoods.


Conclusion
------------
With a low crime rate, an a good housing score, and low fire percentage, we can conclude Shadyside as the best neighborhood to live in.

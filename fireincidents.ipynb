```python
import pandas as pd
import matplotlib.pyplot as plt
chip = pd.read_csv("fires_data.csv")
fire_incidents = chip[chip['type_description'].str.contains('Fire', case=False, na=False)]
shadyside_fire_incidents = fire_incidents[fire_incidents['neighborhood'].str.contains('Shadyside', case=False, na=False)]
total_fires = len(fire_incidents)
shadyside_fires = len(shadyside_fire_incidents)
percentage_fires_shadyside = (shadyside_fires / total_fires) * 100 if total_fires > 0 else 0
print("Total fire incidents in Pittsburgh:", total_fires)
print("Fire incidents in Shadyside:", shadyside_fires)
print("Percentage of fire incidents in Shadyside:", round(percentage_fires_shadyside, 2), "%")
neighborhood_counts = fire_incidents['neighborhood'].value_counts().head(20) 
plt.figure(figsize=(6, 3))
neighborhood_counts.plot(kind='bar', color='firebrick')
plt.title('Top 20 Pittsburgh Neighborhoods by Number of Fire Incidents', fontsize=12)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Fire Incidents', fontsize=8)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

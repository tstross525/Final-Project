```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read file data
filename = pd.read_csv('crimes_neighborhoods.csv')

# 2. Filter 'Shadyside' in the 'INCIDENTLOCATIONS' column
locations = filename['INCIDENTNEIGHBORHOOD']
shadyside_references = locations.str.contains('Shadyside', case=False, na=False)
shadyside_data = filename[shadyside_references]

# 3. How many crimes in Shadyside?
num_crimes = len(shadyside_data)
print("Crimes in Shadyside:", num_crimes)

# How many total crimes?
total_crimes = len(filename)
print("Total crimes:", total_crimes)

# Ratio:
shadyside_crimes_percentage = float(num_crimes / total_crimes)
print(f"The percentage of Pittsburgh crimes that occur in Shadyside is {shadyside_crimes_percentage:.2%}")

# 4. Visual representation 
other_crimes = total_crimes - num_crimes
labels = ['Shadyside', 'Other']
sizes = [num_crimes, other_crimes]

plt.figure(figsize=(10,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Crime Incidents: Shadyside vs. Other Pittsburgh Neighborhoods")
plt.axis('equal')
plt.show()
```
